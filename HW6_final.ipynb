```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mne, glob
from scipy.io import loadmat
```

```python
location = "C:/Users/alexc/OneDrive/Documents/Taiwan Classes/Second Semester/Brain Computer Interfaces/Week 13/BCI-HW6/Matlab Procedure/Dataset/Resulting mat files/"

files8Hz = glob.glob(location+"8Hz.mat")
files14Hz = glob.glob(location+"14Hz.mat")
files28Hz = glob.glob(location+"28Hz.mat")

freq8 = loadmat(files8Hz[0])
freq14 = loadmat(files14Hz[0])
freq28 = loadmat(files28Hz[0])
```

```python

index = []
for i in range(1,61):
    index.append("Trial #" + str(i))

freq8_df = pd.DataFrame(freq8['FFTpower'], index = index)
freq14_df = pd.DataFrame(freq14['FFTpower'], index = index)
freq28_df = pd.DataFrame(freq28['FFTpower'], index = index)
```

```python
freq8_df["Target_Hz"] = "8"
freq14_df["Target_Hz"] = "14"
freq28_df["Target_Hz"] = "28"

freq8_df.head()
#freq14_df.head()
#freq28_df.head()
```

```python
frequencies_df = [freq8_df,freq14_df, freq28_df]

for freq in frequencies_df:
    freq["Channel"] = np.NaN

    freq.Channel.loc[0:20] = "O1"
    freq.Channel.loc[20:40] = "OZ"
    freq.Channel.loc[40:60] = "O2"

freq8_df.head()
```

```python
all_freq_df = pd.concat([freq8_df,freq14_df],join="inner")
all_freq_df = pd.concat([all_freq_df,freq28_df],join="inner")
```

```python
#To rename the index to have 180, and not 3 sets of (1 to 60)
new_index = []
for i in range(1,181):
    new_index.append("Trial #" + str(i))

all_freq_df = all_freq_df.reset_index(drop = True, inplace=False)
all_freq_df = np.array(all_freq_df)
all_freq_df = pd.DataFrame(all_freq_df, index = new_index)
all_freq_df.head()
```

```python
from sklearn.utils import shuffle
all_freq_df = shuffle(all_freq_df)
all_freq_df
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

x = all_freq_df.iloc[:,0:59]
y = all_freq_df.iloc[:,60:61]

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.05,random_state=42)
```

```python
#Import svm model
from sklearn import svm


rbf = svm.SVC(kernel='rbf', gamma=0.5, C=0.1)

rbf.fit(X_train, y_train)

y_pred = rbf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
